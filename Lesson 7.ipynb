# Task 1 Посчитать количество простых чисел
Вводятся n натуральных чисел больше 2 (надо проверить). Посчитать, сколько среди них простых чисел.

```python
n = int(input())
total_simple_number = n

for i in range(n):
    current_num = int(input())
    while current_num <= 2:
        print("Введите число больше двух")
        current_num = int(input())
    
    for i in range(2, int(current_num ** 0.5) + 1):
        if current_num % i == 0:
            total_simple_number -= 1
            break
            
print('Число простых чисел:', total_simple_number)        
    
```

# Task 3 Посчитать общее количество определенных цифр в числах
Посчитать, сколько раз встречается определенная цифра в введенной последовательности чисел. Количество вводимых чисел и цифра, которую необходимо посчитать, задаются вводом с клавиатуры.

```python
n = int(input())
sum = 0
for i in range(1, n + 1):
    cur_num = int(input('Введите число: '))
    cur_digit = int(input('Введите цифру для поиска: '))
    res = 0
    while cur_num != 0:
        if cur_num % 10 == cur_digit:
             res += 1
        cur_num //= 10
    print(res)
    
```

# Task 4 Доказательство гипотезы Сиракуз
Доказать гипотезу Сиракуз на диапазоне чисел. Гипотеза Сиракуз утверждает, что любое натуральное число сводится к единице в результате повторения следующих действий над самим числом и результатами этих действий.
1. Если число четное следует разделить его на 2.
2. Если нечетное, то умножить его на 3, прибавить 1 и разделить на 2.

```python
def syracuse(start, stop):
    for i in range(start, stop + 1):
        print(i, end=' ')
        while i != 1:
            if i % 2 == 0:
                i /= 2
            else:
                i = (i * 3 + 1) / 2
            print(i, end=' ')
        print('-- Theorem Fulfilled')

start, stop = map(int, input().split())
syracuse(start, stop)
```

# Task 6 Лапки и тараканы
У тараканов и пауков вместе n лапки. Сколько могло быть тараканов и пауков, если у таракана - 6 лапок, а у паука – 8.

```python
n = int(input())
result = 0

for i in range(0, n // 6 + 1):
    for j in range(0, 10):
        if 2 * 6 + j * 8 == n:
            result += 1
            
```

```python
%%time
n = int(input())
result = 0
for i in range(0, n // 6 + 1):
    for j in range(0, (n - i * 6) // 8 + 1):
        if i * 6 + j * 8 == n:
            print('Тараканы: {} \nПауки: {}'.format(i, j), end='\n\n')
            result += 1
print('Общее число:', result)
```
