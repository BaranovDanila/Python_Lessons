### Здача 1
Дана последовательность длинной N.

Найти первое вхождение числа x в последовательность, если такого числа нет - вывести -1.

```python
def find_pos(arr, x):
    for i in range(len(arr)):
        if arr[i] == x:
            return i
    return -1


find_pos([1, 2, 3, 4, 1, 4, 5, 1, 5, 123, -4], 1)
```

```python
def find_pos_1(arr, x):
    result = -1
    for i in range(len(arr)):
        if result == -1 and arr[i] == x:
            result = i
    return result


find_pos_1([1, 2, 3, 4, 1, 4, 5, 1, 5, 123, -4], 1)
```

Решим аналогичную задачу - отличие в том, что нам необходимо найти последнее вхождение числа x.

```python
def find_pos_right(arr, x):
    result = -1
    for i in range(len(arr)):
        if arr[i] == x:
            result = i
    return result


find_pos_right([1, 2, 3, 3, 4, 5, 123, 3, -4], 3)
```

```python
find_pos_right('1234343433', '3')
```

### Задача 2 
Необходимо найти максимальный элемент последовательности.

```python
def find_max(arr):
    max_val_id = 0
    for i in range(1, len(arr)):
        if arr[i] > arr[max_val_id]:
            max_val_id = i
    return arr[max_val_id]


def find_max_2(arr):
    max_val = arr[0]
    for i in range(1, len(arr)):
        if arr[i] > max_val:
            max_val = arr[i]
    return max_val
        

print(find_max([1, 2, 3, 3, 4, 5, 123, 3, -4]))
print(find_max_2([1, 2, 3, 3, 4, 5, 123, 3, -4]))
```

### Задача 3
Дана последовательности длины больше, чем 1.

Найти максимальное значение и второе после него значение по величине.

```python
def find_2_max(arr):
    max_1 = max(arr[0], arr[1])
    max_2 = min(arr[0], arr[1])
    
    for i in range(2, len(arr)):
        if arr[i] > max_1:
            max_2 = max_1
            max_1 = arr[i]
        elif arr[i] > max_2:
            max_2 = arr[i]
    return max_1, max_2
    
    
max_1, max_2 = find_2_max([1, 2, 3, 3, 5, 0, 123, 6, 3, -4])
print(f'Самое большое число последовательности {max_1}')
print(f'Самое большое число последовательности после вычёркивания самого большого {max_2}')
```

### Задача 4
Найти все самые короткие слова в последовательности из слов.

```python
def find_min_words(arr):
    result = []
    min_len = len(arr[0])
    for i in range(1, len(arr)):
        if len(arr[i]) < min_len:
            min_len = len(arr[i])
            
    for i in range(len(arr)):
        if len(arr[i]) == min_len:
            result.append(arr[i])
    
    return ' '.join(result)


find_min_words(['abra', 's', 'r', 'k', 'cadabra', 'foo', 'do', 'w', 'fasdf'])
```

```python
def rle_coding(string):
    def pack(s, cnt):
        if cnt > 1:
            return s + str(cnt)
        return s

    result = []
    start_pos = 0
    current_val = string[0]

    for i in range(len(string)):
        if current_val != string[i]:
            result.append(pack(current_val, i - start_pos))
            current_val = string[i]
            start_pos = i
    result.append(pack(current_val, len(string) - start_pos))
    return ''.join(result)


rle_coding('aaaabbbccxyzddddeeefffaaaaabbbbbbbbb')
```
